```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


%matplotlib widget

plt.rcParams["figure.figsize"] = (15,8)



np.random.seed(1)
```

```python
n = 20
x = np.random.uniform(0,1,n)
y = np.random.uniform(0,1,n)
z = np.arange(0,n)
```

```python
points = np.array([x,y,z]).T.reshape(-1,1,3) #shape n x 1 x 3
segments = np.concatenate([points[:-1], points[1:]], axis=1) #shape n-1 x 2 x 3, joins each point to its successor
```

```python
cmap = plt.get_cmap("cool")
colors = [cmap(float(ii)/(n-1)) for ii in range(n-1)]
```

```python

fig = plt.figure()
ax = fig.add_subplot(111,projection="3d")


for ii in range(n-1):
    segii = segments[ii]
    lii, = ax.plot(segii[:, 0], segii[:, 1], segii[:, 2], color=colors[ii], linewidth=2.52)
    lii.set_solid_capstyle("round")


angle=50
ax.view_init(30, angle)
ax.set_zlim(0, max(z))
plt.title("3D-figure")
plt.show()
```
